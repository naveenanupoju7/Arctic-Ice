```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image


img = cv2.imread('G:\TROP ICSU INTERNSHIP\Image processing assignment/SSMI_sea_ice_minimum_w_dates.2020.tif')

plt.figure(figsize=(20,8))
plt.imshow(img) #


grid_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,8))
plt.imshow(grid_RGB) # Printing the original picture after converting to RGB


grid_HSV = cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV) # Converting to HSV
plt.figure(figsize=(20,8))
plt.imshow(grid_HSV) #

#for black colour

lower_black = np.array([0,0,0])
upper_black = np.array([255,255,0])
mask_for_black= cv2.inRange(grid_HSV, lower_black, upper_black)
plt.figure(figsize=(20,8))
plt.imshow(mask_for_black)


#for white colour
lower_white = np.array([0,0,168])
upper_white = np.array([172,111,255])
mask_for_white= cv2.inRange(grid_HSV, lower_white, upper_white)



plt.figure(figsize=(20,8))
plt.imshow(mask_for_white)

print(mask_for_black[1,1])
print(mask_for_white[1,1])
```
